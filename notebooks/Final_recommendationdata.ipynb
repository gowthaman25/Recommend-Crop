```python
merge_crop = pd.read_csv('../Data-raw/MergeFileCrop.csv')
del merge_crop['Unnamed: 0']
reco_fert = merge_fert
```

```python
#Add +/-3 for every NPK value
import random
import numpy as np
temp = pd.DataFrame(columns = ['N','P','K'])
N = []
P = []
K = []

for i in range(0,merge_crop.shape[0]):
    crop = merge_crop.label.iloc[i]
    N1 = reco_fert[reco_fert['Crop'] == crop]["N"].iloc[0] + random.randint(-20,20)
    P1 = reco_fert[reco_fert['Crop'] == crop]["P"].iloc[0] + random.randint(-5,20)
    K1 = reco_fert[reco_fert['Crop'] == crop]["K"].iloc[0] + random.randint(-5,5)
    N.append(N1)
    P.append(P1)
    K.append(K1)
    
temp['N'] = pd.Series(N)
temp['P'] = pd.Series(P)
temp['K'] = pd.Series(K)

temp.head(20)
   
```

```python
merge_crop['N'] = temp['N']
merge_crop['P'] = temp['P']
merge_crop['K'] = temp['K']
```

```python
merge_crop
```

```python
merge_crop = merge_crop[[ 'N', 'P', 'K','temperature', 'humidity', 'ph', 'rainfall', 'label']]
```

```python
merge_crop.to_csv("../Data-processed/crop_recommendation.csv",index=False)
```
